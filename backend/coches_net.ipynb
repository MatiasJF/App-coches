```python
def req_all(url, payload, headers):
    page_num = 1
    total_pages = 1
    res = []

    while page_num <= total_pages:
        payload["pagination"]["page"] = page_num

        response = requests.post(url, json=payload, headers=headers)
        data = response.json()
        res += data["items"]
        total_pages = data["meta"]["totalPages"]
        page_num += 1

    return res

def pay(marca, car_id):
    id = list(car_id.keys())[list(car_id.values()).index(marca)]
    return {"pagination":{"page":1,"size":30},"sort":{"order":"desc","term":"relevance"},"filters":{"isFinanced":False,"price":{"from":0,"to":3000},
    "bodyTypeIds":[],"categories":{"category1Ids":[2500]},"contractId":0,"drivenWheelsIds":[],"environmentalLabels":["B"],"equipments":[],"fuelTypeIds":[],"hasPhoto":None,"hasWarranty":None,
    "hp":{"from":None,"to":None},"isCertified":False,"km":{"from":0,"to":260000},"luggageCapacity":{"from":None,"to":None},"onlyPeninsula":True,"offerTypeIds":[0,2,3,4,5],"provinceIds":[],
    "sellerTypeId":0,"transmissionTypeId":0,"vehicles":[{"make":str(marca),"makeId":id}],"year":{"from":2001,"to":None}}}
```
